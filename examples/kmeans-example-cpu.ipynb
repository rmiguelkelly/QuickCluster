```python
from quickcluster import KMeans

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Create some sample data
N = 1000
(data, _) = make_blobs(N, centers=3)
```

```python
# Initialize and fit the sample data
kmeans = KMeans(3)
kmeans.fit(data)
```

```python
# Plot the data and the computed centroids
data_t = data.T
centers = kmeans.centroid_centers().T

plt.scatter(data_t[0], data_t[1])
plt.scatter(centers[0], centers[1])
```
